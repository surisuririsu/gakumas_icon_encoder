```python
import os
import shutil
```

```python
os.makedirs('data/p_items', exist_ok=True)
for filename in os.listdir('gk-img/docs/p_items/icons'):
    id = os.path.splitext(filename)[0]
    cls_dir = os.path.join('data/p_items', id)
    os.makedirs(cls_dir, exist_ok=True)
    shutil.copy(
        os.path.join('gk-img/docs/p_items/icons', filename),
        os.path.join(cls_dir, 'icon.webp')
    )
```

```python
os.makedirs("data/skill_cards", exist_ok=True)
for filename in os.listdir("gk-img/docs/skill_cards/icons"):
    base = os.path.splitext(filename)[0]
    cls_dir = os.path.join("data/skill_cards", base)
    os.makedirs(cls_dir, exist_ok=True)
    shutil.copy(
        os.path.join("gk-img/docs/skill_cards/icons", filename),
        os.path.join(cls_dir, "icon.webp"),
    )
```
